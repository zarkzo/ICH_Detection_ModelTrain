# split

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
import shutil
from tqdm import tqdm
```

```python
# Baca label file
df = pd.read_csv("data_55k.csv")
```

```python
# Split stratified by "any"
train_df, temp_df = train_test_split(df, test_size=0.20, stratify=df["any"], random_state=42)
test_df, val_df = train_test_split(temp_df, test_size=0.50, stratify=temp_df["any"], random_state=42)
```

```python
print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
```

```python
# misalnya kamu sudah punya dataframe hasil split
# train_df, val_df, test_df
# dan daftar kolom label:
label_cols = ['any', 'subdural', 'intraparenchymal', 'subarachnoid', 'intraventricular', 'epidural']

# hitung distribusi tiap label per split
train_dist = train_df[label_cols].sum()
val_dist = val_df[label_cols].sum()
test_dist = test_df[label_cols].sum()

# tampilkan dalam satu tabel untuk dibandingkan
dist_df = pd.DataFrame({
    'Train': train_dist,
    'Val': val_dist,
    'Test': test_dist
})

# tambahkan proporsi terhadap total data split
dist_df['Train_%'] = dist_df['Train'] / len(train_df)
dist_df['Val_%'] = dist_df['Val'] / len(val_df)
dist_df['Test_%'] = dist_df['Test'] / len(test_df)

print(dist_df)
```

```python
# Simpan CSV hasil split
train_df.to_csv("data_train.csv", index=False)
val_df.to_csv("data_val.csv", index=False)
test_df.to_csv("data_test.csv", index=False)
```

```python
def copy_images(df, src_dir, dst_dir):
    os.makedirs(dst_dir, exist_ok=True)
    for img_id in tqdm(df["ImageID"], desc=f"Copying to {dst_dir}"):
        src = os.path.join(src_dir, f"{img_id}.png")
        dst = os.path.join(dst_dir, f"{img_id}.png")
        if os.path.exists(src):
            shutil.copy2(src, dst)
```
